Dataset of mobility patterns from article: https://www.nature.com/articles/s41597-020-00734-5 - "Multiscale dynamic human mobility flow dataset in the U.S. during the COVID-19 epidemic".

At first, we're analyzing the mobility between the NYC's county, by selecting the data referring to the period from March, 1st and last days of April 2020: in this way simulated data should be coherent with real data of first covid-19 pandemic outbreak of Multiscale dynamic human mobility flow dataset in the U.S. during the COVID-19 epidemic of 2020.

Idea is to change mobilty matrices on the network on which we are simulating the SIR model every week, dividing the average number of weekly commuter for seven days.
