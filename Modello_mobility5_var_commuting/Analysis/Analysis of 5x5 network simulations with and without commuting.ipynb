### Prevalence curves in 100 simulations without and with variations in commuting patterns.

Let's first consider the simulations performed with the SIRcompleto_cluster2 model with parameters $\mu = 1/3 $ and $R_0 = 3$ and then the simulations with the variation of commuting patters in the first 10 weeks of the endemic spreading, performed with SIRcompleto_weekly_mobility code.

Here I present the results for Manhattan borough.

```python
f_prevalence_man={}
f_prevalence_bronx={}
f_prevalence_brook={}
f_prevalence_queens={}
f_prevalence_staten={}

nrun = 100 #numero delle simulazioni

x=0
for i in range(nrun):
    
    path='/home/claudia/Scrivania/Tesi NYC pandemic /Modello_base/FileOutput_R0_3/outputPREVALENCE'
    extens='.txt'
    complete_path = path + str(x) + extens
    input_complete_file = pd.read_csv(complete_path, sep=" ", names=['time step', 'Manhattan'+str(x), 'Bronx'+str(x), 'Brooklyn'+str(x), 'Queens'+str(x),'Staten_Island'+str(x), 'Out'+str(x)])
    input_complete_file.drop(['Out'+str(x)], axis='columns', inplace=True)
    
    f_prevalence_man[x] = pd.DataFrame(input_complete_file[['time step','Manhattan'+str(x)]], columns=['time step','Manhattan'+str(x)])
    f_prevalence_man[x]['Manhattan'+str(x)] = f_prevalence_man[x]['Manhattan'+str(x)].div(1628706)
    
    f_prevalence_bronx[x]= pd.DataFrame(input_complete_file[['time step','Bronx'+str(x)]], columns=['time step','Bronx'+str(x)])
    f_prevalence_bronx[x]['Bronx'+str(x)] = f_prevalence_bronx[x]['Bronx'+str(x)].div(1418207)
    
    f_prevalence_brook[x]= pd.DataFrame(input_complete_file[['time step','Brooklyn'+str(x)]], columns=['time step','Brooklyn'+str(x)])
    f_prevalence_brook[x]['Brooklyn'+str(x)] = f_prevalence_brook[x]['Brooklyn'+str(x)].div(2559903)
    
    f_prevalence_queens[x]= pd.DataFrame(input_complete_file[['time step','Queens'+str(x)]], columns=['time step','Queens'+str(x)])
    f_prevalence_queens[x]['Queens'+str(x)] = f_prevalence_queens[x]['Queens'+str(x)].div(2253858)
    
    f_prevalence_staten[x]=pd.DataFrame(input_complete_file[['time step','Staten_Island'+str(x)]], columns=['time step','Staten_Island'+str(x)])
    f_prevalence_staten[x]['Staten_Island'+str(x)] = f_prevalence_staten[x]['Staten_Island'+str(x)].div(476143)
    
    x+=1;
```

```python
#data from simulations with variations in commuting patterns
f_prevalence_man_comm={}
f_prevalence_bronx_comm={}
f_prevalence_brook_comm={}
f_prevalence_queens_comm={}
f_prevalence_staten_comm={}

x=0
for i in range(nrun):
    
    path2='/home/claudia/Scrivania/Tesi NYC pandemic /Modello_mobility5_var_commuting/FileOutput_R0_3_commutingvar/outputPREVALENCE'
    extens='.txt'
    complete_path2 = path2 + str(x) + extens
    input_complete_file2 = pd.read_csv(complete_path2, sep=" ", names=['time step', 'Manhattan'+str(x), 'Bronx'+str(x), 'Brooklyn'+str(x), 'Queens'+str(x),'Staten_Island'+str(x), 'Out'+str(x)])
    input_complete_file2.drop(['Out'+str(x)], axis='columns', inplace=True)
    
    f_prevalence_man_comm[x] = pd.DataFrame(input_complete_file2[['time step','Manhattan'+str(x)]], columns=['time step','Manhattan'+str(x)])
    f_prevalence_man_comm[x]['Manhattan'+str(x)] = f_prevalence_man_comm[x]['Manhattan'+str(x)].div(1628706)
    
    f_prevalence_bronx_comm[x]= pd.DataFrame(input_complete_file2[['time step','Bronx'+str(x)]], columns=['time step','Bronx'+str(x)])
    f_prevalence_bronx_comm[x]['Bronx'+str(x)] = f_prevalence_bronx_comm[x]['Bronx'+str(x)].div(1418207)
    
    f_prevalence_brook_comm[x]= pd.DataFrame(input_complete_file2[['time step','Brooklyn'+str(x)]], columns=['time step','Brooklyn'+str(x)])
    f_prevalence_brook_comm[x]['Brooklyn'+str(x)] = f_prevalence_brook_comm[x]['Brooklyn'+str(x)].div(2559903)
    
    f_prevalence_queens_comm[x]= pd.DataFrame(input_complete_file2[['time step','Queens'+str(x)]], columns=['time step','Queens'+str(x)])
    f_prevalence_queens_comm[x]['Queens'+str(x)] = f_prevalence_queens_comm[x]['Queens'+str(x)].div(2253858)
    
    f_prevalence_staten_comm[x]=pd.DataFrame(input_complete_file2[['time step','Staten_Island'+str(x)]], columns=['time step','Staten_Island'+str(x)])
    f_prevalence_staten_comm[x]['Staten_Island'+str(x)] = f_prevalence_staten_comm[x]['Staten_Island'+str(x)].div(476143)
    
    x+=1;
```

```python
prevalence_man = pd.DataFrame()
prevalence_bronx = pd.DataFrame()
prevalence_brook = pd.DataFrame()
prevalence_queens = pd.DataFrame()
prevalence_staten = pd.DataFrame()

nrun = 100 #numero delle simulazioni

x=0
for i in range(nrun):
    
    path='/home/claudia/Scrivania/Tesi NYC pandemic /Modello_base/FileOutput_R0_3/outputPREVALENCE'
    extens='.txt'
    complete_path = path + str(x) + extens
    input_file = pd.read_csv(complete_path, sep=" ", names=['time step', 'Manhattan'+str(x), 'Bronx'+str(x), 'Brooklyn'+str(x), 'Queens'+str(x),'Staten_Island'+str(x), 'Out'+str(x)])
    input_file.drop(['Out'+str(x)], axis='columns', inplace=True)
    
    prevalence_man = prevalence_man.join(input_file['Manhattan'+str(x)], how="outer")
    prevalence_man['Manhattan'+str(x)] = prevalence_man['Manhattan'+str(x)].div(1628706)
    
    prevalence_bronx = prevalence_bronx.join(input_file['Bronx'+str(x)], how="outer")
    prevalence_bronx['Bronx'+str(x)] = prevalence_bronx['Bronx'+str(x)].div(1418207)
    
    prevalence_brook = prevalence_brook.join(input_file['Brooklyn'+str(x)], how="outer")
    prevalence_brook['Brooklyn'+str(x)] = prevalence_brook['Brooklyn'+str(x)].div(2559903)
    
    prevalence_queens = prevalence_queens.join(input_file['Queens'+str(x)], how="outer")
    prevalence_queens['Queens'+str(x)] = prevalence_queens['Queens'+str(x)].div(2253858)
    
    prevalence_staten = prevalence_staten.join(input_file['Staten_Island'+str(x)], how="outer")
    prevalence_staten['Staten_Island'+str(x)] = prevalence_staten['Staten_Island'+str(x)].div(476143)
    
    x+=1;

prevalence_man = prevalence_man.dropna()
prevalence_bronx = prevalence_bronx.dropna()
prevalence_brook = prevalence_brook.dropna()
prevalence_queens = prevalence_queens.dropna()
prevalence_staten = prevalence_staten.dropna()
```
